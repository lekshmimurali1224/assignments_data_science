```python
#Importing important libraries
import pandas as pd
import numpy as np
```

```python
#importing the data
beml_df = pd.read_csv('BEML.csv')
beml_df.head()
```

```python
#Importng the second dataset
glaxo_df = pd.read_csv('GLAXO.csv')
glaxo_df.head()
```

```python
#Filtering the required columns
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
#Creating the index
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
beml_df.head()
```

```python
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
glaxo_df.head()
```

```python
#Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
#plt.plot(beml_df.Close)
plt.plot(glaxo_df.index, glaxo_df.Close)
```

```python
#Data Visualization
plt.figure(figsize = (5,3))#Size of graph 5X3
plt.plot(glaxo_df.index, glaxo_df.Close)
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Closing price of GLAXO Stock')
plt.show()
```

```python
#Data Visualization
plt.figure(figsize = (7,2))#Size of graph 7X2
plt.plot(beml_df.index, beml_df.Close)
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Closing price of BEML Stock')
plt.show()
```

```python
#Adding a "gain" column
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
beml_df.head()
```

```python
#Dropping the first row with NaN Value
beml_df.dropna(inplace = True)
beml_df.head()
```

```python
#Adding a "gain" column
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1) #Function per percentage cange
glaxo_df.head()
```

```python
#Dropping the first row with NaN Value
glaxo_df.dropna(inplace = True)
glaxo_df.head()
```

```python
#Plot the gain
plt.figure(figsize = (7,2))#Size of graph 7X2
plt.plot(beml_df.index, beml_df.gain)
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML Stock')
plt.show()
```

```python
#Plot the glaxo gain
plt.figure(figsize = (7,2))#Size of graph 7X2
plt.plot(glaxo_df.index, glaxo_df.gain)
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of GLAXO Stock')
plt.show()
```

```python
#Plot the gains
plt.figure(figsize = (5,3))
plt.plot(beml_df.index, beml_df.gain, c = 'red')
plt.plot(glaxo_df.index, glaxo_df.gain, c ='green')
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML and GLAXO Stock')
plt.show()
```
