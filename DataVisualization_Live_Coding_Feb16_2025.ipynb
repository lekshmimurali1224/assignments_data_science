```python
#Importing important libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatter Plot
#Creating the data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 120, 83, 87, 94, 78, 77, 85]

#Setting the figure size (in inches)
plt.figure(figsize = (5,3))

#Plot
#plt.scatter(x, y, color = 'red')
plt.scatter(x, y, color = 'red', marker = 'o', s = 50)
#Set the x and y values
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
#Show the plot
plt.show()
```

```python
# Line Plot
# Creating the data
year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3, 9.9]

# Plot
plt.figure(figsize = (5,3 ))
plt.plot(year, unemployment_rate, color = 'green', marker = 'o')
#Set the x and y values
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate across years')
#Show the plot
plt.show()
```

```python
#Bar Chart
# Creating the data
#data = {
#    'c' : 20,
#   'c++' : 15,
#    'Java' : 30,
#    'Python' : 35
#    }
x = ['C', 'C++', 'Java', 'Python', 'PHP']
y = [22.2, 17.6, 8.8, 8, 7.7]
#Setting the figure size
plt.figure(figsize=(5,3))
#Creating the bar chart
#plt.bar(data.keys(), data.values(), color = 'blue')
plt.bar(x, y, color = 'blue', width = 0.5)
#Set the x and y values
plt.xlabel('Programming Languages')
plt.ylabel('Nu:Of students')
plt.title('Popularity of Programming Language')
#Show the plot
plt.show()
```

```python
#loading the dataset
df = sns.load_dataset('tips')
df.head()
```

```python
print(df.shape)
```

```python
#Countplot()
#Setting the figure size
plt.figure(figsize = (5,3))
#Plot
sns.countplot(x = 'day', data = df)
plt.show()
```

```python
#Countplot()
#Setting the figure size
plt.figure(figsize = (5,3))
#Plot
sns.countplot(x = 'smoker', data = df)
plt.show()
```

```python
#Boxplot
plt.figure(figsize = (5,3))
plt.boxplot(df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.ylabel('Total bill')
plt.show()
```

```python
#Histogram
plt.figure(figsize = (5,3))
plt.hist(df['total_bill'], bins = 20)
plt.title('Histogram of Total Bill')
plt.xlabel('values')
plt.ylabel('Frequency')
plt.show()
```

```python
#Stacked Bar Chart
import numpy as np

#Creating the data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 2, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 25, 15, 20])

#Plot the bars in stacked manner
plt.bar(x, y1, color = 'green')
plt.bar(x, y2, bottom = y1, color = 'red')
plt.bar(x, y3, bottom = y1 + y2, color = 'blue')
plt.bar(x, y4, bottom = y1 + y2 + y3, color = 'yellow')

#Setting the labels
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Stacked Bar Chart')
plt.legend(['Round 1', 'Round 2', 'Round 3', 'Round 4'])
plt.show()
```

```python
#Pie Chart
#Creating the data
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]

#Plot
plt.figure(figsize = (5,3))
#plt.pie(sizes, labels = labels, autopct = '%.2f%%', startangle = 120) # Will show anticlockwise
plt.pie(sizes, labels = labels, autopct = '%.2f%%')
plt.title('Pie Chart')
plt.show()
```
