# ***Which team has the highest winning percentage?***

```python
# Get a list of all unique teams from the dataset
teams = np.unique(ds2[['team1', 'team2']].values)

# Calculate total matches played by each team (appearing as team1 or team2)
total_matches = ds2['team1'].value_counts() + ds2['team2'].value_counts()
total_matches = total_matches.reindex(teams).fillna(0).astype(int)

# Calculate total wins for each team
total_wins = ds2['winner'].value_counts()
total_wins = total_wins.reindex(teams).fillna(0).astype(int)

# Calculate the winning percentage for each team
winning_percentage = (total_wins / total_matches * 100).fillna(0)

# Find the team with the highest winning percentage
highest_winning_team = winning_percentage.idxmax()
highest_winning_percentage = winning_percentage.max()

# Display the results
print("Winning Percentage of Each Team:")
print(winning_percentage.sort_values(ascending=False))

print(f"\nTeam with the Highest Winning Percentage: {highest_winning_team}")
print(f"Winning Percentage: {highest_winning_percentage:.2f}%")
```

# ***Is there any lucky venue for a particular team?***

```python
games_played = ds2.groupby(['venue', 'team1']).size() + ds2.groupby(['venue', 'team2']).size()
games_played = games_played.reset_index(name='total_games')

# Calculate wins for each team at each venue
wins_per_venue = ds2.groupby(['venue', 'winner']).size().reset_index(name='wins')

# Merge total games played with wins
venue_stats = pd.merge(games_played, wins_per_venue, left_on=['venue', 'team1'], right_on=['venue', 'winner'], how='left')

# Fill NaN values with 0 in wins
venue_stats['wins'] = venue_stats['wins'].fillna(0)

# Calculate win percentage for each team at each venue
venue_stats['win_percentage'] = (venue_stats['wins'] / venue_stats['total_games']) * 100

# Define a threshold for "lucky" (e.g., win rate > 70%)
lucky_venues = venue_stats[venue_stats['win_percentage'] > 70]

# Display the results
print("Teams with high win percentages at certain venues (potentially lucky venues):")
print(lucky_venues[['venue', 'team1', 'win_percentage']].sort_values(by='win_percentage', ascending=False))
```

# ***Who has the highest strike rate?***

```python
print(ds1.columns)
```

```python
# Filter out balls that are wides, as they are not counted as faced by the batsman
valid_balls = ds1[ds1['extras_type'] != 'wides']

# Group by batsman to calculate total runs and balls faced
batsman_stats = valid_balls.groupby('batsman').agg(
    runs_scored=pd.NamedAgg(column='batsman_runs', aggfunc='sum'),
    balls_faced=pd.NamedAgg(column='ball', aggfunc='count')
).reset_index()

# Calculate strike rate
batsman_stats['strike_rate'] = (batsman_stats['runs_scored'] / batsman_stats['balls_faced']) * 100

# Find the player with the highest strike rate
highest_sr_player = batsman_stats.loc[batsman_stats['strike_rate'].idxmax()]

highest_sr_player
```
